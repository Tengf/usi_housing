```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
import csv
import dateutil
%pylab inline
```

```python
complaints= pd.read_csv('data_cleaned_2015_MN/complaints_2015.csv') 
violations= pd.read_csv('data_cleaned_2015_MN/violations_2015.csv') 
#gap= pd.read_csv('data_cleaned_2015_MN/c_v_a_gap_2015.csv')
```

```python
complaints_time = complaints.groupby(['ReceivedDate']).count()
violations_time = violations.groupby(['InspectionDate']).count()
complaints_time = complaints_time['ComplaintID']
violations_time = violations_time['ViolationID']
```

```python
#dataframe to array
C_array = complaints_time.as_matrix()
V_array = violations_time.as_matrix()
#print C_array
print len(C_array)
print len(V_array)
```

```python
date_list_c = complaints_time.index.tolist()
date_list_v = violations_time.index.tolist()
```

```python
mylist_c = list(xrange(365))
mylist_v = list(xrange(361))
month_list = ['Jan', 'Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

##Timeseries
* What happend in October (very high peak)
* Summer vs. Winter ?
* Obvious cycle (weekly)
* Time lag between complaints and violations

```python
fig = plt.figure(figsize=(18, 6))
plt.ylabel('#of Complaints & Violations', fontsize=10)
plt.xlabel('Month (2015)', fontsize=10)
#plt.xlim(1994,2013)
#plt.ylim(-4,4.5)
locs = np.arange(365,step=30.5) # locations
plt.xticks(locs, month_list, rotation=-45 )
plt.plot(mylist_c,C_array, alpha=1, linewidth=0.5, c='Red')
plt.plot(mylist_v,V_array, alpha=0.5, linewidth=0.3, c='Blue')
```
